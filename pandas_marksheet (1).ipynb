```python
numeric_cols =['Math_Score','English_Score','Science_Score','Attendance']
orgininal_stats={}
```

```python
for col in numeric_cols:
    mean_val= df[col].mean()
    df[col]=df[col].fillna(mean_val)
    std_val=df[col].std()
    orgininal_stats[col]=(mean_val,std_val)
    df[col]=df[col].clip(lower=mean_val-3*std_val,upper=mean_val+3*std_val)#normal data set to find lower value 
    print(f"clipped values in {col}")
```

```python
for col in numeric_cols:
    mean_val=df[col].mean()
    std_val=df[col].std()
    mean_val,std_val=orgininal_stats[col]
    lower_bound=mean_val-3*std_val
    upper_bound=mean_val+3*std_val
    outliers_remaining=df[(df[col]<lower_bound)|(df[col]>upper_bound)].shape[0]
    if outliers_remaining==0:
        print(f"outliers in {outliers_remaining} in {col}")
    else:
        print("outliers are there")
```

```python
for col in numeric_cols:
    mean_val,std_val=orgininal_stats[col]
    lower_bound=mean_val-3*std_val
    upper_bound=mean_val+3*std_val
    outliers_remaining=df[(df[col]<lower_bound)|(df[col]>upper_bound)].shape[0]
    print(f"outliers in {outliers_remaining} to {col}")
```
